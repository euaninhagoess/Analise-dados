Colunas existentes na planilha


```python
colunas_feitas = [
    'PAY_AMT1','PAY_AMT2','PAY_AMT3','PAY_AMT4','PAY_AMT5','PAY_AMT6'
]
```

Colunas que ja foram examinadas

```python
colunas_restantes = [
    col for col in df.columns
    if (('BILL' in col) and col not in colunas_feitas)
]
```

Destacando quais colunas financeiras ainda não foram examinadas

```python
colunas_restantes
```

```python
df[colunas_restantes].describe()
```

A análise das variáveis de valor da fatura (BILL_AMT) evidencia uma elevada dispersão nos dados, observada pela presença de valores significativamente superiores à média, caracterizando possíveis outliers. Além disso, foram identificados valores negativos, que podem estar relacionados a créditos, estornos ou ajustes financeiros nas contas dos clientes.

```python
import matplotlib.pyplot as plt
```

```python
df[colunas_restantes].hist(bins=20, figsize=(15,8), layout=(2,3))
```

```python
df[colunas_feitas].describe()
```

As estatísticas das variáveis de valor de pagamento (PAY_AMT) indicam que uma parcela significativa dos clientes não realiza pagamentos conforme o esperado, evidenciado pela grande quantidade de valores iguais a zero identificados na análise. Além disso, observa-se a presença de valores extremamente elevados, que influenciam e distorcem as médias, caracterizando a existência de outliers no conjunto de dados. Esses resultados demonstram a alta variabilidade do comportamento de pagamento entre os clientes.

```python
df[colunas_feitas].hist(
    bins=20,
    figsize=(12, 8),
    layout=(2, 3),
    xrot=50
)
```

```python
(df[colunas_feitas] == 0).sum()
```

Com base nos resultados obtidos, a análise apresenta coerência em relação ao que foi observado anteriormente nos histogramas. Conforme identificado, uma parcela significativa dos clientes não realiza o pagamento de suas dívidas, resultando em altos índices de inadimplência. Esse comportamento é evidenciado pela forte concentração de valores próximos ou iguais a zero nos histogramas, indicando baixa frequência de pagamentos em diversos registros analisados.

```python
import numpy as np
```

```python
pagamento_nao_zero = df[colunas_feitas].replace(0, np.nan)
```

```python
pagamento_log = pagamento_nao_zero.apply(np.log10)
```

```python
pagamento_log.hist(
    bins=20,
    figsize=(12,8),
    layout=(2,3),
    xrot=45
)

plt.tight_layout()
plt.show()
```
